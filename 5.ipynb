1. **Unconditional Storage:** No post-selection of data points.
2. **Beating the No-Cloning Limit:** Achieving fidelity $F > 50\%$ for unknown coherent states.
3. **Scalability:** Operating at high temperature (~80°C) without cryogenics.

##The Quantum Motivation & Physical Mechanism

### 1. Why do we need a Quantum Memory?
In classical systems, a memory is essentially a buffer which simply measures a signal, retains the information for a certain time, and then re-broadcasts it. However, in the quantum regime, we face the **No-Cloning Theorem**:
* We cannot measure an unknown quantum state without collapsing it.
* **The Classical Limit:** For a coherent state $|\alpha\rangle$ with a random phase, the best possible classical "measure-and-prepare" strategy yields a fidelity of exactly 0.5.
* **The Challenge:** To prove the memory is quantum, the authors demonstrate $F > 0.5$ across all input states without post-selection.

---

### 2. The Atomic Interface ($^{87}$Rb)
The authors utilize a three-level $\Lambda$-type atomic system to map light onto matter.

* **The Levels:** $|1\rangle$ (Initial ground), $|2\rangle$ (Storage ground), and $|3\rangle$ (Excited bridge).
* **The Raman Coupling:** A strong Coupling Beam ($\Omega_c$) and a weak Signal Field ($\mathcal{E}$) drive a two-photon transition.
* **Effective Two-Level Approximation:** Because the detuning $\Delta$ is large (far-detuned regime), the excited state $|3\rangle$ is never significantly populated, preventing spontaneous emission noise.
---

### 3. Gradient Echo Memory (GEM) Protocol
The "Magic" of GEM lies in the spatial mapping of frequency to the length of the cell using an external magnetic field gradient:
$$\delta(z,t) = \eta(t) z$$

1. **Absorption:** Each frequency component of the pulse finds a unique $z$-position in the cell where it is in resonance, creating a **Spin Wave** (collective atomic coherence).
2. **Storage:** The coupling beam is turned off, "freezing" the coherence in the ground states.
3. **Reversal & Echo:** To retrieve, we flip the magnetic gradient ($-\eta$). The atomic phases now start neutralizing, and the pulse is re-emitted as an optical echo.

## Input Signal and Evolution in the Memory Cell

The memory system can be considered as $\Lambda$-GEM process. This setup, under a weak probe, off resonance is equivalent to two level system. For this maxwell-bloch equations are solved in weak probe and far detuned regimes. Consider a three level system with photon detuning $\Delta$, a classical coupling beam with Rabi frequency of $\Omega_c$ and a weak quantum signal field $\hat{E}$. The two photon detuning $\delta(z,t)$ is given as $\delta(z,t)= \eta(t)z$.

image.png


The equations are -
$$\begin{aligned}
\dot{\hat{\sigma}}_{13} &= -(\gamma + \gamma_0/2 + i\Delta)\hat{\sigma}_{13} + ig\hat{\mathcal{E}} + i\Omega_c\hat{\sigma}_{12} + \hat{F}_{13} \\
\dot{\hat{\sigma}}_{12} &= -(\gamma_0 + i\delta(z,t))\hat{\sigma}_{12} + i\Omega_c^*\hat{\sigma}_{13} + \hat{F}_{12} \\
\frac{\partial}{\partial z} \hat{\mathcal{E}} &= i\mathcal{N}\hat{\sigma}_{13}
\end{aligned}$$

where $\mathcal{N} = \frac{gN}{c}$ is the effective linear atomic density and $N$ is the total number of atoms in the interaction volume.

The Langevin operators $\hat{F}_{13}$ and $\hat{F}_{12}$ account for noise arising from spontaneous emission $\gamma$ and ground state decoherence $\gamma_0$ respectively. In a linear regime, where we assume the probe is very weak, these terms only add noise to the state that is required to maintain the Heisenberg uncertainty between conjugate observables. We ignore the Langevin noise terms in our modelling, in this regime, and add vacuum noise to the observables to correct for the efficiency of the memory. Also, the interaction time is small compared to the time between collisions of atom, ignore the population redistribution terms, so that the atoms prepared in state $|1\rangle$ don't change.

Starting with all population in state $|1\rangle$ initially, and with assumption of adiabaticity, we have characteristic timescale, $T$, much slower than the decay rate ($\gamma$), $1/T \ll \gamma$. Additionally, in off-resonance regime, the one-photon detuning significantly exceeds the spontaneous emission rate ($\Delta \gg \gamma$).

By integrating the Maxwell equation for the probe field with the motion equation $\hat{\sigma}_{13}$, it can be concluded that for atomic densities greater than unity, the excited state can be effectively eliminated if $1/dT \ll \gamma$ and $\Delta \gg \gamma d$ (where $d$ represents optical depth). These conditions arise from the collective interaction between the many-atom ensemble and the optical fields. Finally, by treating the coupling beam amplitude as a real value and maintaining a non-zero detuning ($\Delta \neq 0$), the original three equations can be condensed into a simplified, effective two flavour form, in a frame where speed of light in medium is normalised by refractive index, and coupling frequency matches light shift term. The effective equations are:
$$\begin{aligned}
\dot{\hat{\sigma}}_{12} &= -(\gamma_{12} + i\delta'(z,t))\hat{\sigma}{12} - i g' \mathcal{E} \\
\frac{\partial}{\partial z}\mathcal{E} &= i\mathcal{N}' \hat{\sigma}_{12}
\end{aligned}$$
where $g' = \frac{g \Omega_c}{\Delta}$, and $\mathcal{N}' = \frac{\mathcal{N}\Omega_c}{\Delta}$


These equations can be solved in fourier domain which are given by
$$\begin{aligned}
\dot{\tilde{\sigma}} &= -\gamma_{12}2\pi \delta(k) + \eta \frac{\partial \Sigma}{\partial k} - \frac{ig'\mathcal{N}'}{k}\tilde{E} \\
\tilde{E} &= \frac{\mathcal{N}'}{k}\tilde{\sigma}
\end{aligned}$$
The Fourier domain representation introduces additional computational overhead, such as the $\partial \Sigma / \partial k$ term and potential singularities at $k=0$. Consequently, we perform the numerical simulation in the original time-space domain, which provides a more straightforward and stable implementation.

This numerical implementation is executed through:
- We use RK4 scheme to integrate the Maxwell-Bloch equations in the real-space domain. The simulation tracks the evolution of the atomic coherence ($\hat{\sigma}_{12}$) and the probe field ($\hat{\mathcal{E}}$) over a discretized grid ($N_z, N_t$). The storage and retrieval mechanism is driven by a linear magnetic field gradient that defines the detuning $\delta(z, t)$
- Consistent with the two-level and three-level memory models, the retrieval of the stored excitation is achieved through time-reversal of the detuning slope. By flipping the gradient ($\eta \to -\eta$) at the midpoint of the simulation, the atomic dipoles, which had previously dephased, begin to rephase. This constructive interference results in an echo pulse that exits the medium in the forward direction. The simulation treats this echo as a time-reversed copy of the input, mapping the storage process to a symmetric retrieval process
- We then map the transition from field operators to discrete quadrature values by doing a temporal mode integration. The complex amplitudes of the input and output fields are projected onto normalized temporal envelopes:$$\alpha = \int \hat{\mathcal{E}}(t) u^*(t) dt$$where $u(t)$ is the mode function defining the gaussian pulse duration.
- The complex amplitudes are then scaled by the mean photon number ($\bar{n}$) and mapped to measurable quadratures ($X_\theta$). We then project the integrated signal onto a local oscillator with varying phase, in balanced homodyne detector $$X_\theta = 2\text{Re}(\alpha e^{-i\theta}) + \delta X_{vac}$$
- By introducing stochastic Gaussian noise ($\delta X_{vac}$) with unit variance, we generate a distribution of quadrature values that accounts for vacuum fluctuations.
- Choosen values of free parameter, as it is not mentioned in paper, $g' = 0.1$, $\mathcal{N} = 10$, $\gamma = 0.05$ and $\eta = 5.0$, which also fulfils the constraint that optical depth should be greater than 1. We take a gaussian beam as input coherent pulse.

```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import coherent, fock_dm, ket2dm #, wigner, Qobj
from qutip import tensor, destroy, identity, expect, coherent, coherent_dm
import qutip as qt
from scipy.optimize import curve_fit
from scipy.special import factorial
from scipy.special import eval_hermite, factorial
from numpy.linalg import norm
```

```python
def run_simulation(nbar, N_samples=100000, plot=True, *, Nz=800, Nt=2000, L=1.0, T=10.0, seed=None):
    # Grid
    dz = L / Nz
    dt = T / Nt
    z = np.linspace(0, L, Nz)
    t = np.linspace(0, T, Nt)

    # Parameters
    g = 0.1
    N = 10.0
    gamma = 0.05
    eta_grad = 5.0
    delta_z = eta_grad * z

    t0 = 2.0
    sigma_t = 0.5

    def input_pulse(t_):
      return np.exp(-(t_ - t0) ** 2 / (2 * sigma_t ** 2))

    E = np.zeros(Nz, dtype=np.complex128)
    sigma12 = np.zeros(Nz, dtype=np.complex128)
    E_in = np.zeros(Nt, dtype=np.complex128)
    E_out = np.zeros(Nt, dtype=np.complex128)
    def sigma_rhs(sigma, E_val, delta_val):
        return -(gamma + 1j * delta_val) * sigma - 1j * g * E_val

    for n in range(Nt - 1):
        delta = delta_z if n < Nt // 2 else -delta_z
        k1 = dt * sigma_rhs(sigma12, E, delta)
        k2 = dt * sigma_rhs(sigma12 + 0.5 * k1, E, delta)
        k3 = dt * sigma_rhs(sigma12 + 0.5 * k2, E, delta)
        k4 = dt * sigma_rhs(sigma12 + k3, E, delta)
        sigma12_next = sigma12 + (k1 + 2 * k2 + 2 * k3 + k4) / 6.0
        E_next = np.empty_like(E)
        E_next[0] = input_pulse(t[n + 1])
        if Nz > 1:
            E_next[1:] = E_next[0] + dz * 1j * N * np.cumsum(sigma12_next[1:])
        E = E_next
        sigma12 = sigma12_next
        E_in[n + 1] = E[0]
        E_out[n + 1] = E[-1]

    # Modes
    u_in = input_pulse(t)
    u_in /= np.sqrt(np.trapezoid(np.abs(u_in) ** 2, t))

    u_out = input_pulse(T - t)
    u_out /= np.sqrt(np.trapezoid(np.abs(u_out) ** 2, t))

    alpha_in_raw = np.trapezoid(np.conj(u_in) * E_in, t)
    alpha_out_raw = np.trapezoid(np.conj(u_out) * E_out, t)
    eta_eff = (np.abs(alpha_out_raw) ** 2) / (np.abs(alpha_in_raw) ** 2 + 1e-12)
    phase_correction = np.angle(alpha_in_raw) - np.angle(alpha_out_raw)
    alpha_out_raw = alpha_out_raw * np.exp(1j * phase_correction)

    # Phase sampling
    rng = np.random.default_rng(seed)
    theta = rng.uniform(-np.pi, np.pi, N_samples)

    def alpha_scale(nbar_, alpha):
        return np.sqrt(nbar_) * alpha / (np.abs(alpha_in_raw) + 1e-12)

    alpha_in = alpha_scale(nbar, alpha_in_raw)
    alpha_out = alpha_scale(nbar, alpha_out_raw)

    # LO phase shift
    lo_shift = np.exp(-1j * np.pi / 2)
    alpha_in *= lo_shift
    alpha_out *= lo_shift

    # Homodyne signals
    mu_in = 2 * np.real(alpha_in * np.exp(-1j * theta))
    mu_out = 2 * np.real(alpha_out * np.exp(-1j * theta))

    variance = 1
    X_in = mu_in + rng.normal(0, np.sqrt(variance), N_samples)
    X_out = mu_out + rng.normal(0, np.sqrt(variance), N_samples)

    if plot:
        fig, axs = plt.subplots(1, 2, figsize=(12, 5))
        alphas = [alpha_in, alpha_out]
        X_datas = [X_in, X_out]
        titles = ['Input', 'Output']

        for i in range(2):
            ax = axs[i]
            X_data = X_datas[i]
            alpha = alphas[i]
            amp = 2 * np.abs(alpha)
            phase_plot = np.angle(alpha)
            ax.scatter(theta, X_data, s=1, alpha=0.03, color='navy')
            theta_line = np.linspace(-np.pi, np.pi, 200)
            X_line = amp * np.cos(theta_line - phase_plot)
            ax.plot(theta_line, X_line, color='orangered', linewidth=2, linestyle='--')
            peak_theta = (phase_plot + np.pi) % (2 * np.pi) - np.pi
            ax.axhline(amp, color='black', linestyle='--', alpha=0.6)
            ax.axhline(-amp, color='black', linestyle='--', alpha=0.6)
            label = f"$S_{{{'in' if i == 0 else 'out'}}} = {2 * amp:.2f}$"
            ax.annotate(
                label,
                xy=(peak_theta + 1.5, amp),
                xytext=(peak_theta + 1.5, amp + 1.5),
                arrowprops=dict(arrowstyle="->", color='red'),
                ha='center',
                fontsize=11,
            )
            ax.set_title(titles[i])
            ax.set_xlabel("Phase (rad)")
            if i == 0:
                ax.set_ylabel("Normalized quadrature amplitude")
            ax.set_xlim(-np.pi, np.pi)
            ax.set_ylim(-7, 7)
            ax.set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
            ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$'])
        plt.tight_layout()
        plt.show()

    return {"X_in": X_in, "X_out": X_out, "theta": theta, "eta_eff": eta_eff}
```

```python
sim = run_simulation(nbar=3.4)
```

## Reconstruction of Density Matrix

The grand goal is to determine the quantum state of the light that was stored in the quantum memory,and compare it with the input state sent into the memory.

Classically, everything can be measured directly. However, in quantum mechanics, we cannot directly observe the full state.

Thus instead, we measure outcomes and reconstruct the state. The quantity of our interest is the density matrix $\hat{\rho}$.

### What is a density matrix?

The density matrix represents our knowledge (or lack of knowledge) of a quantum state. It represents a statistical mixture of probable states, taking into account their associated probabilities.

A general quantum state can be written as:

$$
\hat{\rho} = \sum_{m,n} \rho_{mn}|m\rangle \langle n|
$$

This is in the Fock basis: $\rho_{nn}$ is the probability of having $n$ photons, while $\rho_{nm}$ $(n\neq m)$ describes quantum coherence

The goal therefore, is to find all $\rho_{nm}$

### Homodyne Measurement

We start from the observable given by

$$
\hat{X}_\theta = \hat{a}e^{-i\theta} + \hat{a}^\dagger e^{i\theta}
$$

where $\hat{a}$ and $\hat{a}^\dagger$ are anhilation and creation operators and $\theta$ is the phase that can be controlled experimentally.
Measurement gives $x\in \mathbb{R}$, thus we employ the following protocol:

For every $\theta_i \in (-\pi, \pi)$, we create multiple interval bins in the quadrature space. Through several measurements, the counts in these bins for every $\theta_i$ are measured.

These are used to create an experimental discrete probability distribution $P_\text{exp}(x|\theta)$.

Each Homodyne measurement corresponds to the measurement operator

$$
\hat{\Pi}(\theta) = |x_\theta\rangle \langle x_\theta |
$$

where $|x_\theta\rangle$ is the eigenstate of the $\hat{X}_\theta$ operator. Also, we have

$$
|x_\theta\rangle = e^{i\theta \hat{a}^\dagger \hat{a}}|x\rangle
$$

where $|x\rangle$ is the eigenstate for $\theta = 0$. In Fock basis,

$$
|x\rangle = \sum_n \psi_n(x)|n\rangle
$$

Expanding $|x_\theta\rangle$, we get

$$
|x_\theta \rangle = \sum_{n}\psi_n(x)e^{-in\theta}|n\rangle
$$

where $\psi_n(x)$ is the $n^\text{th}$ Hermite-Gaussian function. Building the projector in Fock basis, we get

$$
\hat{\Pi}(x,\theta) = \sum_{n,m}\psi_n(x)\psi_m(x)e^{i(m-n)\theta}|n\rangle \langle m|
$$

Now, the probability distribution is related to the density matrix as follows:

$$
P(x|\theta) = \text{Tr}\left(\hat{\rho}\hat{\Pi}(x,\theta)\right) = \sum_{n,m}\rho_{nm}\psi_n(x)\psi_m(x)e^{i(n-m)\theta}
$$

### MLE Algorithm

Let $\theta_j$ denote the phase bins and $x_k$ denote the quadratue bins. Then:

$$
N_{jk} = \text{Number of counts in bin}\ (x_k, \theta_j)
$$

The experimental probability is calculated as:

$$
P_\text{exp}(x_k|\theta_j) = \frac{N_{jk}}{\sum_k N_{jk}}
$$

Theoretically,

$$
P_\text{th}(x_k|\theta_j) = \text{Tr}\left(\hat{\rho}\hat{\Pi}_{jk}\right)\ \ \ \text{where}\ \ \ \hat{\Pi}_{jk} = |x_k,\theta_j\rangle \langle x_k, \theta_j|
$$

Likelihood function assuming independent measurements:

$$
\mathcal{L}(\hat{\rho}) = \Pi_{j,k}P_\text{th}(x_k|\theta_j)^{N_{jk}}
$$

Taking log on both sides,

$$
\log \mathcal{L} = \sum_{j,k}N_{jk}\log \text{Tr}\left(\hat{\rho}\hat{\Pi}_{jk}\right)
$$

The following constraints have to be enforced: $\hat{\rho}\geq 0$ and $\text{Tr}\hat{\rho} = 1$. Thus, the problem turns into a variational problem. We maximize:

$$
\log \mathcal{L}(\hat{\rho}) - \lambda\left(\text{Tr}\hat{\rho} - 1\right)
$$

In order to maximize the above, we implement the following iterative algorithm:

#### Step 1: Initialize

$$
\hat{\rho}^{(0)} = \frac{1}{N}\mathbb{I}_{N\times N}
$$

where $N$ is the cut-off for the number of Fock states in the Hilbert space (taken to be 20 in our case).

#### Step 2: Compute Probabilities

$$
p_{jk} = \text{Tr}\left(\hat{\rho}\hat{\Pi}_{jk}\right)
$$

#### Step 3: Build R

$$
R = \sum_{j,k}\frac{N_{jk}}{p_{jk}}\hat{\Pi}_{jk}
$$

#### Step 4: Update

$$
\hat{\rho} \rightarrow \frac{R\hat{\rho}R}{\text{Tr}\left(R\hat{\rho}R\right)}
$$

```python
def reconstruct_mle(phases,quads,*,hilbert_size=20,iterations=30,n_phi_bins=30,n_x_bins=120,x_range=None,eps=1e-12,seed=None,verbose=False):
    phases = np.asarray(phases, dtype=np.float64)
    quads = np.asarray(quads, dtype=np.float64)
    phases = (phases + np.pi) % (2 * np.pi) - np.pi
    if x_range is None:
        q_lo, q_hi = np.quantile(quads, [0.001, 0.999])
        pad = 0.1 * (q_hi - q_lo + eps)
        x_range = (q_lo - pad, q_hi + pad)

    phi_edges = np.linspace(-np.pi, np.pi, n_phi_bins + 1)
    x_edges = np.linspace(x_range[0], x_range[1], n_x_bins + 1)
    counts, _, _ = np.histogram2d(phases, quads, bins=[phi_edges, x_edges])
    counts = counts.astype(np.float64)
    total_counts = counts.sum()

    phi_centers = 0.5 * (phi_edges[:-1] + phi_edges[1:])
    x_centers = 0.5 * (x_edges[:-1] + x_edges[1:])

    a = qt.destroy(hilbert_size)
    x_op = (a + a.dag())
    x_evals, x_kets = x_op.eigenstates()
    x_evals = np.asarray(x_evals, dtype=np.float64)
    x_idx = np.abs(x_evals[:, None] - x_centers[None, :]).argmin(axis=0)

    # Precompute projectors Pi_{jk} = |x_k,phi_j><x_k,phi_j| in numpy form
    n = np.arange(hilbert_size, dtype=np.float64)
    n_bins = n_phi_bins * n_x_bins
    Pi = np.zeros((n_bins, hilbert_size, hilbert_size), dtype=np.complex128)
    N_flat = counts.reshape(-1)
    b = 0
    for phi in phi_centers:
        phase = np.exp(1j * phi * n)
        for xi in x_idx:
            ket = x_kets[int(xi)].full().reshape(-1)
            rot = ket * phase
            Pi[b] = np.outer(rot, rot.conj())
            b += 1

    # Initial state: maximally mixed
    rho = np.eye(hilbert_size, dtype=np.complex128) / hilbert_size
    for it in range(iterations):
        probs = np.real(np.einsum('ab,kba->k', rho, Pi))
        probs = np.clip(probs, eps, None)
        w = N_flat/probs
        R = np.einsum('k,kab->ab', w, Pi) / total_counts
        R = 0.5 * (R + R.conj().T)
        rho = R @ rho @ R
        rho = 0.5 * (rho + rho.conj().T)
        tr = np.trace(rho)
        rho = rho / (tr + 0j)
        if verbose and (it % 5 == 0 or it == iterations - 1):
            evals = np.linalg.eigvalsh(rho)
            if evals.min() < -1e-8:
                print(f"iter {it}: min eigenvalue {evals.min():.2e}")
    return qt.Qobj(rho, dims=[[hilbert_size], [hilbert_size]])
```

```python
def plot_density_matrices(rho_in, rho_out, N_cut):
    rho_in_abs = np.abs(rho_in)
    rho_out_abs = np.abs(rho_out)

    vmax = max(np.max(rho_in_abs), np.max(rho_out_abs))

    fig = plt.figure(figsize=(12, 5))

    # Create grid with middle space for colorbar
    gs = fig.add_gridspec(1, 3, width_ratios=[1, 0.05, 1])

    ax1 = fig.add_subplot(gs[0])
    cax = fig.add_subplot(gs[1])  # middle colorbar axis
    ax2 = fig.add_subplot(gs[2])

    im1 = ax1.imshow(rho_in_abs, origin='lower', cmap='RdBu_r', vmin=0, vmax=vmax)
    ax1.set_title("Input Density Matrix")
    ax1.set_xlabel('m')
    ax1.set_ylabel('n')

    im2 = ax2.imshow(rho_out_abs, origin='lower', cmap='RdBu_r', vmin=0, vmax=vmax)
    ax2.set_title("Output Density Matrix")
    ax2.set_xlabel('m')
    ax2.set_ylabel('n')

    # Ticks
    for ax in [ax1, ax2]:
        ax.set_xticks(range(0, N_cut, 2))
        ax.set_yticks(range(0, N_cut, 2))

    # Colorbar in the middle
    fig.colorbar(im1, cax=cax, label='|ρ_nm|')

    plt.tight_layout()
    plt.show()
```

```python
sim = run_simulation(nbar=3.4)
theta = sim["theta"]
X_in = sim["X_in"]
X_out = sim["X_out"]
rho_initial = reconstruct_mle(theta, X_in, hilbert_size=20, iterations=25, n_phi_bins=30, n_x_bins=120)
rho_final = reconstruct_mle(theta, X_out, hilbert_size=20, iterations=25, n_phi_bins=30, n_x_bins=120)

rho_in = rho_initial.full()
rho_out = rho_final.full()
plot_density_matrices(rho_in, rho_out, N_cut=20)
```

## Photon number distribution

In quantum optics, the photon number distribution describes the probability of finding $n$ photons in a given quantum state of light. It is defined as:

$$ P(n) = \langle n|\rho|n\rangle $$

where $\rho$ is the density matrix of the optical state and $|n\rangle$ are the Fock (number) states. Thus, the photon number distribution corresponds to the diagonal elements of the density matrix in the Fock basis.


### Coherent states
For a coherent state $|\alpha\rangle$, the photon number distribution follows a Poisson distribution:

$$ P(n) = \frac{\lambda^{n}e^{-\lambda}}{n!}$$

where $\lambda = |\alpha|^{2}$ is the mean number of photons in the state.

Key properties:

* Mean = Variance = $|\alpha|^{2}$
* Represents classical-like light
* No excess noise beyond quantum fluctuations


In the experiment, the input states are near-coherent, so the measured distributions are expected to closely follow Poissonian statistics.


### Effect of Quantum Memory on Photon Statistics

An ideal quantum memory should preserve the photon statistics and not introduce additional noise.

However, real systems may introduce Gaussian noise in the quadratures.

From the supplementary material:
* The memory is modeled as adding equal noise to both quadratures
* This noise modifies the reconstructed photon distribution


### Reconstruction from Homodyne Measurements

The photon number distribution is not measured directly. Instead, the process is:
* Perform homodyne measurements of quadratures
* Reconstruct the density matrix $\rho$ using Maximum Likelihood
* Extract: $$ P(n) = \langle n|\rho|n\rangle $$





```python
def poisson(n, lamb):
  return (lamb**n * np.exp(-lamb)) / factorial(n)
```

### Wigner function representation

The Wigner function provides a phase-space representation of a quantum state, analogous to a probability distribution in classical mechanics.

It is defined as a quasi-probability distribution: $W(x, p)$ over:

$x$: amplitude quadrature

$p$: phase quadrature

Unlike classical probability distributions it can take negative values. These negativities are signatures of non-classicality

Thus, the Wigner function gives an intuitive picture of a quantum state in phase space.

For a density matrix $\rho$, the Wigner function is given by:

$$ W(x, p) = \frac{1}{\pi}\int_{-\infty}^{\infty} \langle x-y|\rho| x+y\rangle e^{2ipy}\,dy$$

Although the Wigner function is not directly measurable, it can be reconstructed via Homodyne detection of quadratures.

The Wigner function correctly reproduces measurable distributions:

$$ P(x) = \int W(x, p) \,dp$$
$$ P(p) = \int W(x, p) \,dx$$

For a coherent state, the Wigner distribution is a Gaussian in phase space



```python
xvec = np.linspace(-5, 5, 100)

W_initial = qt.wigner(rho_initial, xvec, xvec)
W_final = qt.wigner(rho_final, xvec, xvec)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Input state
cont1 = axes[0].contourf(xvec, xvec, W_initial, 100, cmap='RdBu_r')
axes[0].set_title("Input Wigner Function")
axes[0].set_xlabel("x")
axes[0].set_ylabel("p")
fig.colorbar(cont1, ax=axes[0])

# Output (reconstructed) state
cont2 = axes[1].contourf(xvec, xvec, W_final, 100, cmap='RdBu_r')
axes[1].set_title("Reconstructed Wigner Function")
axes[1].set_xlabel("x")
axes[1].set_ylabel("p")
fig.colorbar(cont2, ax=axes[1])

plt.tight_layout()
plt.show()
```

```python
#functions
def get_quadratures(X, theta, tol=0.1):
  def angle_diff(a, b):
    return np.abs((a - b + np.pi) % (2*np.pi) - np.pi)
  mask_x = angle_diff(theta, 0) < tol
  mask_p = angle_diff(theta, np.pi/2) < tol
  x = X[mask_x]
  p = X[mask_p]
  return x, p

def loss_variance(V_meas, eta_det):
    return (V_meas - 1)/eta_det + 1

def compute_TV(Vin_x, Vin_p, Vout_x, Vout_p, eta):
    T_plus  = eta / (1 + Vout_x - Vin_x)
    T_minus = eta / (1 + Vout_p - Vin_p)
    Vcv_plus  = (1 - T_plus) * Vout_x
    Vcv_minus = (1 - T_minus) * Vout_p
    return T_plus, T_minus, Vcv_plus, Vcv_minus
```

```python
nbar_list = [0.67, 3.41, 12, 16.3, 22.4]
results = []

for nbar in nbar_list:
    sim = run_simulation(nbar=nbar, N_samples=100000, plot=False)
    theta = sim["theta"]
    X_in = sim["X_in"]
    X_out = sim["X_out"]
    eta = 0.593
    x_in, p_in = get_quadratures(X_in, theta)
    x_out, p_out = get_quadratures(X_out, theta)

    Vin_x = loss_variance(np.var(x_in), eta)
    Vin_p = loss_variance(np.var(p_in), eta)

    Vout_x = loss_variance(np.var(x_out), eta)
    Vout_p = loss_variance(np.var(p_out), eta)

    hilbert_size = int(nbar + 5*np.sqrt(nbar))
    hilbert_size = max(hilbert_size, 20)

    rho_in = reconstruct_mle(theta, X_in, hilbert_size=hilbert_size)
    rho_out = reconstruct_mle(theta, X_out, hilbert_size=hilbert_size)

    F = qt.fidelity(rho_in, rho_out)**2
    T_plus, T_minus, Vcv_plus, Vcv_minus = compute_TV(Vin_x, Vin_p, Vout_x, Vout_p, eta)

    results.append({"nbar": nbar,"F": F,"Tsum": T_plus + T_minus,"Vprod": Vcv_plus * Vcv_minus })

    print(f"nbar={nbar}")
    print("Vin:", Vin_x, Vin_p)
    print("Vout:", Vout_x, Vout_p)
    print("Tsum:", T_plus + T_minus)
    print("Vprod:", Vcv_plus * Vcv_minus)
```

```python
def fidelity_plot(results):
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    markers = ['o', 's', '^', 'D', '*']
    ax = axs[0]
    for i, r in enumerate(results):
        ax.scatter(r["nbar"], r["F"],marker=markers[i % len(markers)],s=90,facecolor='white',edgecolor='red',linewidth=1.8,zorder=3)
        ax.annotate(f'{r["F"]:.3f}', xy=(r["nbar"], r["F"]), xytext=(0, 8), textcoords='offset points', ha='center', fontsize=9)
    ax.axhline(0.68, linestyle='--', color='purple', linewidth=1.5)
    ax.axhline(0.5, linestyle='--', color='purple', linewidth=1.5)
    ax.text(22, 0.7, "(i)", fontsize=10)
    ax.text(22, 0.52, "(ii)", fontsize=10)
    ax.set_xlim(0, 30)
    ax.set_ylim(0, 1.05)
    ax.set_xlabel("Photon number")
    ax.set_ylabel("Fidelity")
    ax.set_title("a", loc='left', fontweight='bold')
    ax.tick_params(direction='in')

    ax = axs[1]
    x = np.linspace(1, 2, 300)
    ax.fill_between(x, 0, 1,color='#6A7FDB',alpha=0.4,zorder=0)
    ax.axvline(1, linestyle='--', color='purple', linewidth=1.5)
    ax.axhline(1, linestyle='--', color='purple', linewidth=1.5)
    for i, r in enumerate(results):
        ax.scatter(r["Tsum"], r["Vprod"],marker=markers[i % len(markers)],s=90,facecolor='white',edgecolor='red',linewidth=1.8,zorder=3)

    ax.annotate("No-cloning region",xy=(1.2, 0.7), xytext=(1.35, 1.6),arrowprops=dict(arrowstyle="->", color='blue'),color='blue')
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2.2)
    ax.set_xlabel(r"$T^+ + T^-$")
    ax.set_ylabel(r"$V_{cv}^+ \times V_{cv}^-$")
    ax.set_title("b", loc='left', fontweight='bold')
    ax.tick_params(direction='in')
    plt.tight_layout()
    plt.show()
```

The Wigner function of the output state remains approximately Gaussian, which demonstrates that the memory preserves the overall phase-space structure of the quantum state


### Fidelity as a quantitative measure

To rigorously quantify memory performance, we computed the state fidelity between input and output:

$$ F = (Tr(\sqrt{\sqrt{\rho_{in}}\rho_{out}\sqrt{\rho_{in}}}))^{2}$$

This provides a complete measure of similarity between quantum states.

Here, it exceeds the no-cloning limit of 0.68, indicating that the memory operates in the quantum regime.


### T-V Analysis

To further validate the quantum performance, we reconstructed the T–V diagram, as shown in Figure 4 of the paper. We use:
* Transfer coefficient $T$ measures signal preservation
* Conditional variance $V_{CV}$ measures added noise

Two key benchmarks define the quantum regime:
$$ T^{+} + T^{-} \geq 1$$
$$ V_{CV}^{+} * V_{CV}^{-} \leq 1$$

Our results show that the system operates within or near the no-cloning regime
indicating performance beyond classical limits.
